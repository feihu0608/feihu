## 数据预处理(API)

```python
# 1. 删除列
dataset = train_dataset.remove_columns(['cat'])
print(dataset)
```

```python
# 2. 过滤行
dataset = dataset.filter(lambda x:x['review'] is not None)
```

```python
print(dataset)
```

```python
# 3. 划分数据集
dataset = dataset.cast_column("label",ClassLabel(names=['negative','positive']))
dataset_dict = dataset.train_test_split(test_size=0.2,stratify_by_column='label')
print(dataset_dict)
```

## 与分词器配合使用,编码数据

```python
# 获取训练数据集
train_dataset = dataset_dict['train']
print(train_dataset)
```

```python
print(train_dataset[:5])
```

```python
print(train_dataset[10000:10005]['review'])
```

```python
# 定义分词器
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('../data/mybert')
```

```python
# 定义编码函数
def encode(batch):
    inputs = tokenizer(
        batch['review'],
        padding='max_length',
        max_length=64,
        truncation=True,
        # return_tensors='pt',
    )
    inputs['labels'] = batch['label']
    return inputs
```

```python
train_dataset = train_dataset.map(encode, batched=True, remove_columns=['label', 'review'])
```

```python
print(train_dataset)
```

```python
print(train_dataset[:5])
```

```python
dataset_dict = dataset_dict.map(encode, batched=True, remove_columns=['label', 'review'])
```

```python
print(dataset_dict)
```

## 集成DataLoader

```python
from torch.utils.data import DataLoader
```

```python
# 加载文件，创建数据集
train_dataset = load_from_disk('data/train')
```

```python
train_dataset.set_format(type='torch')
```

```python
print(train_dataset[:5])
```

```python
# 创建数据加载器
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
for batch in train_loader:
    print(batch)
    for k, v in batch.items():
        print(k, "->", v.shape)
    break
```
