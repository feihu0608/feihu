## 远程加载

## API调用

```python
text = "我在超级自然语言学习"
```

```python
# 1. 分词
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
# 2. token转换成id
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

```python
# 3. id转换成token
tokens = tokenizer.convert_ids_to_tokens(ids)
print(tokens)
```

```python
# 4. 编码
ids = tokenizer.encode(text)
print(ids)
```

```python
text_pair = "我爱抽奖深度学习"
ids = tokenizer.encode(text,text_pair)
print(ids)
```

```python
# 5. 解码
sentence = tokenizer.decode(ids,skip_special_tokens=True) # 跳掉特殊token
print(sentence)
print(sentence.replace(" ",""))
```

```python
# 6. __call__方法,直接构建模型需要的所有输入参数
inputs = tokenizer(text)
print(inputs)
```

```python
print(inputs["input_ids"])
```

```python
inputs = tokenizer(text,text_pair,padding='max_length',max_length=20,truncation=True)
print(inputs)
```

## 分词器配合模型使用

```python
from transformers import AutoTokenizer,AutoModel,AutoModelForSequenceClassification
```

```python
# 1. 准备原始语料数据
texts = ['我爱自然语言处理', '我爱深度学习', '我爱机器学习']
```

## 基准模型

```python
# 2. 加载模型分词器
model_name_or_path = "../data/mybert"
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)
model = AutoModel.from_pretrained(model_name_or_path)
```

```python
# 3. 编码: 将文本转换为模型输入
inputs = tokenizer(texts,padding=True,return_tensors="pt")
print(inputs)
```

```python
# 4. 模型前向传播
outputs = model(
    input_ids = inputs["input_ids"],
    token_type_ids = inputs["token_type_ids"],
    attention_mask = inputs["attention_mask"],
)
print(outputs)
```

## 带任务头的模型

```python
# 加载模型
model4sc = AutoModelForSequenceClassification.from_pretrained(model_name_or_path)
```

```python
# 编码,得到模型输入
inputs = tokenizer(texts, return_tensors="pt",padding=True)
print(inputs)
```

```python
# 在输入中,添加目标标签
import torch
labels = torch.randint(2,(3,))
inputs["labels"] = labels
print(inputs)
```

```python
# 前向传播
outputs = model4sc(**inputs)
print(outputs)
```

```python
# 反向传播，计算梯度
outputs.loss.backward()
```
