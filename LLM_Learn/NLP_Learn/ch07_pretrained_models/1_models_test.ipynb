## 模型的前向传播

### 基准模型

```python
import torch
```

```python
# 定义输入
vocab_size = 1000
N = 32
L = 11
input_ids = torch.randint(vocab_size, size=(N, L))
print(input_ids)
```

```python
output = model(input_ids)
```

```python
print(output)
```

```python
print(output.last_hidden_state.shape) # 最后一层的隐藏状态(词级分类)
```

```python
print(output.pooler_output.shape)  # pooler_output 是对 last_hidden_state 的第一个token（[CLS] token）的向量进行进一步处理后的结果，专门用于句子级别的分类任务
```

## 带任务头的模型

```python
output = model4sc(input_ids)
print(output)
```

```python
print(output.logits.shape) # 逻辑得分,三分类问题
```

```python
# 传入目标值,自动计算损失值
target = torch.randint(3, size=(N,))
print(target)
```

```python
output = model4sc(input_ids,labels=target)
```

```python
print(output)
```

```python
print(output.loss)
```

```python
output.loss.backward() # 反向传播
```

```python
print(output.loss)
```
