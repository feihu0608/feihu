```python
import math
import torch
```

```python
torch.set_printoptions(precision=6,sci_mode=False) # 保留6位小数并且不使用科学计数法
torch.manual_seed(42) # 写死随机数因子
```

```python
L = 4
dk = 64
```

```python
# 定义 Q K V 矩阵
Q = torch.randn(L,dk,requires_grad=True) # 保留梯度
K = torch.randn(L,dk)
V = torch.randn(L,dk)
```

```python
# 1. 计算注意力评分
scores = Q @ K.T / math.sqrt(dk) # 避免评分差距过大softmax后导致其他信息丢失
print(scores)
```

```python
# 2. 计算注意力权重
weights = torch.softmax(scores, dim = -1)
print(weights)
```

```python
# 3. 计算上下文向量
vectors = weights @ V
print(vectors)
```

```python
print(vectors.shape)
```

```python
# 模拟损失,反向传播
vectors.sum().backward()
```

```python
print(Q.grad)
```
