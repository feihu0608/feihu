## 自行训练词向量

```python
from gensim.models import Word2Vec
# 准备语料
sentences = ["我每天乘坐地铁上班","我每天乘坐公交上班"]
import jieba
sentences = [jieba.lcut(sent) for sent in sentences]
print(sentences)
```

```python
# 构建模型并且训练
model = Word2Vec(
    sentences=sentences, # 已分词的句子序列
    vector_size=10, # 词向量维度
    window=2, # 上下文窗口大小
    min_count=1, # 最小词频(低于则忽略)
    sg=1, # 1 是sikp_gram 0 是CBOW 默认0
    workers=4, # 并行训练线程数
)
```

```python
print(model) # alpha是学习率
```

```python
print(model.wv)
```

```python
print(model.wv["地铁"])
```

```python
print(model.wv.similarity("地铁","公交"))
```

## 真实语料

```python
import pandas as pd
```

```python
# 1.读取数据
df = pd.read_csv("../data/online_shopping_10_cats.csv",encoding='utf-8',usecols=['review']).dropna()
df.head()
```

```python
#2.准备数据:分词
# df["review"] = df["review"].astype(str).fillna("")
sentences = [ [word for word in jieba.lcut(sentence) if word.strip() != ''] for sentence in df["review"]]
# sentences = [ [token for token in jieba.lcut(review) if token.strip() != ''] for review in df["review"]]
print(len(sentences))
```

```python
print(sentences[:5])
```

```python
#3. 定义训练模型
model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1,
)
```

```python
# 查看词向量
print(model.wv["地铁"])
```

```python
# 计算相似度
print(model.wv.similarity("地铁","火车"))
```
