```python
import jieba
```

```python
text = "小明毕业于北京大学计算机系"
```

# 精确模式

```python
generator = jieba.cut(text)
```

```python
for word in generator:
    print(word)
```

```python
words = jieba.lcut(text)
print(words)
```

```python
generator = jieba.cut(text, cut_all=True) # 全模式
for word in generator:
    print(word)
```

```python
words = jieba.lcut(text, cut_all=True)
print(words)
```

## 搜素引擎模式

```python
generator = jieba.cut_for_search(text)
for word in generator:
    print(word)
```

```python
words = jieba.lcut_for_search(text)
print(words)
```
