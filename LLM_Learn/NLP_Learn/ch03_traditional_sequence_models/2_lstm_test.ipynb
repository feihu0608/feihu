```python
import torch
import torch.nn as nn
```

```python
# 定义输入数据
vocab_size = 2000
N = 2
L = 4
input_ids = torch.randint(vocab_size, size=(N, L))
# 词嵌入层
embedding = nn.Embedding(num_embeddings=vocab_size, embedding_dim=3)
# 嵌入层前向传播,得到词向量的列表
input = embedding(input_ids)
print(input.shape)
print(input)
```

```python
# 定义LSTM层
lstm = nn.LSTM(3,5,num_layers=2,bidirectional=True,batch_first=True,proj_size=4) # proj_size 投影参数主要是记忆单元投影隐藏输出时的维度
```

```python
# 前向传播
out_put,(h_n,c_n) = lstm(input)
```
