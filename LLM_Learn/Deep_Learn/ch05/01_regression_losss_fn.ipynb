```python
import torch
import torch.nn as nn
output = torch.randn(5)  # 神经网络的输出: 损失函数的输入
target = torch.randn(5) # 目标值: 正确标签

print(output)
print(target)
```

## MAE

```python
loss_fn = nn.L1Loss()
loss = loss_fn(output, target)
print(loss)
```

## MSE

```python
loss_fn = nn.MSELoss()
loss = loss_fn(output, target)
print(loss)
```

## Smooth l1

```python
loss_fn = nn.SmoothL1Loss()
loss = loss_fn(output, target)
print(loss)
```
