```python
import torch
import torch.nn as nn
from torch.nn import BCEWithLogitsLoss
```

## 二元交叉熵损失函数

```python
# 1. 定义5条数据: 模拟的是神经网络的输出
output = torch.randn(5)
# 2. 定义标签
target = torch.tensor([0, 0, 1, 1, 0]).float()
```

```python
# 损失函数
loss_fn = nn.BCELoss()

# 神经网络的输出, 通过sigmoid映射成1的概率
proba = torch.sigmoid(output)

# 计算损失
loss = loss_fn(proba, target)
print(loss)
```

```python
# 定义损失: 二元交叉熵+sigmoid
loss_fn = BCEWithLogitsLoss()
loss = loss_fn(output, target)
print(loss)
```

## 交叉熵损失函数

```python
output = torch.randn(8, 6)  # 6分类问题
target = torch.tensor([1, 0, 5, 4, 2, 3, 0, 5])
```

```python
# 定义损失函数: softmax+交叉熵
loss_fn = nn.CrossEntropyLoss()

# 计算损失
loss = loss_fn(output, target)
print(loss)
```

```python
# 标签是独热编码
target_onehot = torch.zeros(8, 6)
target_onehot[torch.arange(8), target] = 1

loss = loss_fn(output, target_onehot)
print(loss)
```
