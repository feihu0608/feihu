```python
import  torch
import torch.nn as nn
x = torch.tensor([0.7, -1.2, 2.5, 0, -3.1])
```

## sigmoid

```python
## 1. 直接调用torch的sigmoid函数
y = torch.sigmoid(x)
print(y)
```

```python
## 2. 直接调用tensor的sigmoid函数
y = x.sigmoid()
print(y)
```

```python
## 3. nn模块定义的Sigmoid类
sigmoid = nn.Sigmoid()
print(sigmoid.forward(x))
print(sigmoid(x))
```

## Tanh双曲正切函数

```python
print(torch.tanh(x))
print(x.tanh())
print(nn.Tanh()(x))
```

## ReLu

```python
print(torch.relu(x))
print(x.relu())
print(nn.ReLU()(x))
```

## Softmax

```python
print(torch.softmax(x, dim=-1))
print(x.softmax(dim=-1))
print(nn.Softmax(dim=-1)(x))
```

## 其他激活函数

```python
print(nn.LeakyReLU()(x))
print(nn.PReLU()(x))
print(nn.RReLU()(x))
print(nn.ELU()(x))
print(nn.SiLU()(x))
print(nn.Softplus()(x))
print(nn.GELU()(x))

```
