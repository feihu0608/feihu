```python
import torch
import torch.nn as nn
```

## 全连接层

```python
linear = nn.Linear(3, 4)
```

```python
print(linear.weight)
print(linear.bias)
```

## 向前传播

```python
x = torch.tensor([[10,20, 30], [30, 40, 50]]).float()
y = linear.forward(x)  # (*, out_features)   (2, 4)
print(y)
```

```python
y = linear(x)  # 向前传播
print(y)
```
