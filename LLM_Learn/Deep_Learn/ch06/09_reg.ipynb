```python
import torch
import torch.nn as nn
from torch.optim import SGD
```

## BN层

```python
x = torch.randint(0, 10, (5, 3)).float()
print(x)
```

```python
bn = nn.BatchNorm1d(num_features=3)
```

```python
y = bn(x)
print(y)
```

## 权值衰减

```python
SGD(weight_decay=0.2)
```

## Dropout

```python
x = torch.randint(1, 10, (10, )).float()
print(x)
```

```python
dp = nn.Dropout(p = 0.5)
y = dp(x)
print(y)
```
