```python
import torch
```

## 元素类型的转换

```python
t1 = torch.tensor([10, 20])
print(t1.dtype)
```

```python
# type 方法
t2 = t1.type(torch.float32)
print(t2.dtype)
```

```python
# to 方法
t3 = t1.to(torch.float32)
print(t3.dtype)
```

```python
# 直接调用类型相关的方法
print(t1.double().dtype)
```

## tensor => ndarry

```python
import numpy as np
```

```python
# 内存共享
t1 = torch.tensor([10, 20])
n1 = t1.numpy()
print(n1)
print(type(n1))

t1[0] = 100
print(t1)
print(n1)
```

```python
# 不想共享内存
n1 = t1.numpy().copy()
t1[0] = 200
print(n1)
```

# ndarray => tensor

```python
n1 = np.array([100, 200, 300])

# torch.tensor(n1)
t1 = torch.from_numpy(n1)  # 内存共享
n1[0] = 10
print(t1)
```

```python
# 不想共享内存
t1 = torch.from_numpy(n1).clone()
n1[0] = 50
print(t1)
```

## 张量=> 标量

```python
t1 = torch.tensor([[[[10]]]])  # tensor只有一个元素的情况可以转
print(t1.item())
```
