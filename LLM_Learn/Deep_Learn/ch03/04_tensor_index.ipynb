```python
import torch

t = torch.randint(0, 10, (2, 5, 4))
print(t)
```

## 简单索引

```python
print(t[0][1][1])
print(t[0, 1, 1])
```

```python
print(t[1][1])
```

## 范围索引

```python
print(t[1][1:])
```

```python
print(t[-1, 1:4, 0:3:2])
```

## 列表索引

```python
list1 = [0, 1, 1]
list2 = [1, 2, 3]
list3 = [0, 0, 0]
print(t[list1, list2, list3])
```

```python
list1 = [[0], [1]]
list2 = [1, 2, 3]
print(t[list1, list2])
```

## 布尔索引

```python
# 取出所有的大于5的值
mask = t > 5
#print(mask)
print(t[mask])
print(t[t > 5])
```

```python
# 取出满足条件的行: 首元素大于5
mask = t[:, :, 0] > 5
print(mask)
print(mask.shape)
print(t[mask])
```

```python
# 取出满足条件的列: 首元素大于5
mask = t[:, 0, :] > 5
print(mask)
print(t.mT[mask].mT)
```

```python
# 满足符合条件的矩阵
mask = t[:, 1, 2] > 5
print(mask)
print(t[mask])
```
