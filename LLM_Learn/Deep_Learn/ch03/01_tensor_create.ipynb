## 1. 基于内容创建张量

```python
# 传入一个标量
t1 = torch.tensor(1)
print(t1)
print(t1.shape)
```

```python
# 传入一个列表
t1 = torch.tensor([1, 2])
print(t1)
print(t1.shape)
```

```python
# 传入一二维列表
t1 = torch.tensor([ [10, 20], [30, 40] ])
print(t1)
print(t1.shape)
```

```python
# 传入高维数组
t1 = torch.tensor([ [[10, 20, 10], [30, 40, 30]],  [[10, 20, 40], [30, 40, 50]]])
print(t1)
print(t1.shape)
```

```python
# 传入numpy的ndarry
import numpy as np
t1 = torch.tensor(np.array([[10, 20, 10], [30, 40, 30]]))
print(t1)
print(t1.shape)
```

```python
# 数据类型:  基于内存创建的tensor, 数据类型默认是保持与传入的内容不变
t1 = torch.tensor([1, 2], dtype=torch.float)
print(t1.dtype)
```

## 4. 根据数据区间创建张量

```python
# [start, end)
t1 = torch.arange(start=10, end=100, step=10)
print(t1)
print(t1.shape)
```

```python
# [start, end]
t1 = torch.linspace(10, 100, 5)  # step = (end - start)/ (steps - 1)
print(t1)
print(t1.shape)
```

```python
t1 = torch.logspace(1, 3, 4)  # 10^1 10^2 10^3
print(t1)
print(t1.shape)
```

```python
t1 = torch.logspace(1, 3, 4, base=np.e)  # 10^1 10^2 10^3
print(t1)
print(t1.shape)
```

## 5. 根据填充创建张量


```python
t1 = torch.ones(3, 4, 5)
print(t1)
```

```python
t2 = torch.ones_like(t1)
print(t2)
```

```python
t1 = torch.zeros(2,3)
print(t1)
```

```python
t1 = torch.empty(2,3)
print(t1)
```

```python
t1 = torch.full((2,3), fill_value=20)
print(t1)
```

```python
t1 = torch.eye(3)
print(t1)
```

```python
t1 = torch.eye(3, 4)
print(t1)
```

## 创建随机张量

```python
# 均匀分布:  [0, 1)
t1 = torch.rand(2,3)
print(t1)
```

```python
# 均匀分布
t1 = torch.randint(2, 10, (2,3))
print(t1)
```

```python
# 标准正态分布
t1 = torch.randn(2, 3)
print(t1)
```

```python
# 正态分布
t1 = torch.normal(5, 1, (2,3))
print(t1)
```

```python
# 洗牌  [0, 10) 随机排序
t1 = torch.randperm(10)
print(t1)
```

```python
# 查看随机数种子
print(torch.random.initial_seed())
```
