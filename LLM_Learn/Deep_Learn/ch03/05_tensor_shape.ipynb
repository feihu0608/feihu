```python
import torch

t1 = torch.randint(1, 10, (3, 2, 6))
print(t1)
```

## 变换维度

```python
## 转置 本质: 下标交换   1,0,2 => 2,0,1
print(t1.T)
print(t1.mT)  # 只转置最后2个维度
```

```python
# 交换任意2个维度  (1,0,3) => (0,1,3)
t1.transpose(0, 1)
```

```python
# 重新排列所有维度
print(t1.permute(2, 1, 0))
```

## 形状变化

```python
## reshape
print(t1.reshape(-1))
print(t1.reshape(1, 2, 3, 6))
```

```python
## view
print(t1.view(-1))
print(t1.view(1, 2, 3, 6))
```

```python
t2 = t1.mT
```

```python
print(t2.reshape(-1))
```

```python
#print(t2.view(-1))
print(t1.is_contiguous())
print(t2.is_contiguous())
t2 = t2.contiguous()
print(t2.view(-1))
```

## 增删维度

```python
t1 = torch.randint(1, 10, (2, 4))
print(t1)
```

```python
## 添加维度
t2 = t1.unsqueeze(dim=2)
print(t2.shape)
print(t2)
```

```python
# 删除维度  删除为1的那些维度
print(t2.squeeze())
```

```python
t3 = t1.reshape(1, 1, 1, 1, 2, 1, 4)
print(t3)
print(t3.squeeze().shape)
```

## 张量的拼接

```python
t1 = torch.randint(1, 10, (2, 1, 3))
t2 = torch.randint(1, 10, (2, 3, 3))

print(t1)
print(t2)
```

```python
## 拼接  除了堆叠的维度, 其他要一致
t3 = torch.cat((t1, t2), dim=1)
print(t3)
```

```python
## stack
t1 = torch.randint(1, 10, (2, 3, 3))
t2 = torch.randint(1, 10, (2, 3, 3))
t3 = torch.stack((t1, t2), dim=2)  # 先unsqueeze , 然后再cat
print(t3.shape)
```
