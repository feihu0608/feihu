```python
import torch
```

```python
t1 = torch.tensor([[10, 20, 30], [40, 50, 60]])
```

## 基本运算

```python
# +
print(t1 + 10)
print(t1.add(10)) # t1不变,返回一个新的tensor
print(t1.add_(10))  # 修改t1
print(t1)
```

## 元素级运算

```python
t1 = torch.tensor([[1, 2, 3], [4, 5, 6]])
t2 = torch.tensor([[10, 20, 30], [40, 50, 60]])
```

```python
# 张量+张量  对位相加   形状一样或者满足关闭
print(t1 + t2)
```

```python
print(t1 * t2)  # 哈达玛积
print(t1.mul(t2))
```

## 矩阵相乘

```python
t1 = torch.tensor([[1, 2, 3], [4, 5, 6]])
t2 = torch.tensor([[10, 30], [40, 60] , [40, 60]])
```

```python
print(t1 @ t2)
print(t1.mm(t2))
print(t1.matmul(t2))
```

```python
t1 = torch.tensor([[[1, 2, 3], [4, 5, 6], [6, 5, 4]], [[3, 2, 1], [3, 2, 1], [3, 2, 1]]])
t2 = torch.tensor([[[1, 2], [3, 4], [5, 6]], [[6, 5], [4, 3], [2, 1]]])

print(t1.shape, t2.shape)
```

```python
# 最后2个维度满足 矩阵相乘, 其他保持一致
print(t1 @ t2)
# print(t1.mm(t2))  # 只能是矩阵
print(t1.matmul(t2))
```

## 统计运算

```python
t1 = torch.randint(1, 10, (2, 3, 4))
print(t1)
```

```python
# 1. sum
print(t1.sum())
print(t1.sum(dim=1))
```

```python
# 2. mean
print(t1.float().mean(dim=0))
```

```python
# 3. max/min
print(t1.float().max(dim=0))
```

```python
# 4. var
print(t1.float().var(dim=0))
```

```python
# 5. 去重
print(t1.unique())
```

```python
# 6. 排序: 默认是矩阵的每一行单独排序  ==== dim=2
print(t1.sort(dim=0))
```
