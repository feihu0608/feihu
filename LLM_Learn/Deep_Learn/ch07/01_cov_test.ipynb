```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# 1. 读取图片
img = plt.imread("/Users/lzc/remote_project/dl_20260331 /data/duck.jpg")
print(img.shape)  # (h, w, c)  => (c, h, w)

data = torch.tensor(img).permute(2, 0, 1).float()
```

```python
# 2. 卷积层
conv = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, stride=1, padding=1, bias=True)
```

```python
# 3. 做卷积(向前传播)
out_data = conv(data)
print(out_data.shape)
```

```python
# 显示图片
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(img)

out_img = out_data.permute(1, 2, 0).detach().int()
out_img = torch.clamp(out_img, 0, 255).numpy()
ax[1].imshow(out_img)
plt.show()
```
