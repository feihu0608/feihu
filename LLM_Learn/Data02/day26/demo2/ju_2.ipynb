# 常用方法

```python
# 相关系数
# corr()
# 最终取值为  -1 ~ 1 之间
# 1 正相关 A涨B也涨
# 0 无相关
# -1 负相关 A涨B跌

import pandas as pd

s1 = pd.Series([1,2,3])
s2 = pd.Series([5,6,7])
print(s1.corr(s2)) # 正相关


```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([7,6,5])
print(s1.corr(s2)) # 负相关
```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([1,-1,1,-1])
print(s1.corr(s2))
```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([500,6500,70000])
print(s1.corr(s2))
```

```python
# 协方差
# 正 一起涨一起跌
# 正 方向相反

s1 = pd.Series([1,2,3])
s2 = pd.Series([500,6500,70000])
print(s1.cov(s2))
```

```python
s1 = pd.Series([1,2,3])
s2 = pd.Series([70000,6500,500])
print(s1.cov(s2))
```

```python
# 绘制直方图 看一组数据 每个数据的分布状态
s1 = pd.Series([1,1,2,2,2,3,4,5,6,7,7,7,7])
s1.hist() # bins 指定柱子的数量 如果不指定 则按照默认的去重元素个数
```

```python
# items() 获取到索引及其对应的值
# 可以迭代遍历

s1 = pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
for i , v in  s1.items():
    print(i,v)
```
