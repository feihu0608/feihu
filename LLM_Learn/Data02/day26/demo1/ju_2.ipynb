# Seris的属性

```python
# index 属性
import pandas as pd

s1 = pd.Series([11,22,33,44,55],index=['a','b','c','d','e'],name="numbers")
s1
```

```python
# f Field 字段 属性的意思
print(s1.index)
```

```python
# index属于可迭代对象 所以直接使用for循环遍历
for i in s1.index:
    print(i)
```

```python
# values 属性
print(s1.values)
print(type(s1.values))
```

```python
# ndim 维度
# field 和 property 这两种都可以理解为是类中的属性
# field表示未封装的  property 是封装之后
print(s1.ndim) # Seris是一维的数据结构
```

```python
# 形状
print(s1.shape) # 一列行数据
```

```python
# 元素个数
print(s1.size)
```

```python
# 元素类型
print(s1.dtype)
print(s1.dtypes)
```

```python
# 名字属性 name
print(s1.name)
```

```python
# 按照标签切片 或者 单个元素操作
print(s1.loc["a"])
print(s1.loc["b"])
print(s1.loc["c"])
print(s1.loc["d"])
```

```python
# 切片 根据标签切片 注意：包括起始和结束
s1.loc["a" : "d"]
```

```python
# 通过iloc 隐式索引 进行切片或者获取单个元素
s1 = pd.Series([11,22,33,44,55],index=['a','b','c','d','e'],name="numbers")

print(s1.iloc[0])
print(s1.iloc[1])
print(s1.iloc[2])
print(s1.iloc[3])
```

```python
# 切片 跟原来规则一致  左闭右开
s1.iloc[0:3]
```

```python
# at 获取单个元素 注意：at不能切片
# at 按照标签获取单个元素
print(s1.at["a"])
```

```python
# iat 按照隐式索引 获取单个元素
print(s1.iat[0])
print(s1.iat[1])
```
