# DataFrame 运算

```python
import pandas as pd
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] }
)
print(df)
```

```python
print(df * 2)
```

```python
df1 = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] }
)

df2 = pd.DataFrame(
    data = {"id" : [101,102,103],"name" :["赵四","广坤","小宝"],"age" :[22,23,34] }
)


print(df1 + df2)
```

```python
df1 = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] }
)

df2 = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] }
)

print(df1 / df2) # 报错
print(df1 - df2) # 报错
```
