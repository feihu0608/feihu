# DataFrame 常用函数

```python
import pandas as pd

df = pd.DataFrame({"A" : [2,5,3,7,4],"B" : [1,6,2,8,3]})
df
```

```python
print(df.cumprod()) #累计乘积  不写参数 跟 axis = 0 效果相同
```

```python
print(df.cumprod(axis=1)) #累计乘积  axis =1 表示横向操作
```

```python
df = pd.DataFrame({"A" : [2,5,3,7,4],"B" : [1,6,2,8,3]})

print(df)
```

```python
# 一阶差分 用当前行减去上一行的值 如果为第一行 则显示NaN
print(df.diff())
```

```python
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] }
)
print(df)
```

```python
print(df.sort_index()) # 按照默认索引进行排序
```

```python
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] },
    index = ["a","c","d","b"]
)
print(df.sort_index())
```

```python
# 按照年龄排序
# ascending 不写 或者 ascending=True 表示升序
# ascending = False 表示降序
print(df.sort_values(by="age", ascending=False))
```

```python
# 获取age列最大的前2条数据
print(df.nlargest(n=2, columns="age"))
```

```python
print(df.nsmallest(n=1, columns="age")) # 获取年龄列 前1个最小的数据
```
