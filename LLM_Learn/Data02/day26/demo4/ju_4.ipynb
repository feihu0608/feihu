# DataFrame 修改操作

```python
import pandas as pd


df = pd.DataFrame({"age": [20, 30, 40, 10], "name": ["张三", "李四", "王五", "赵六"], "id": [101, 102, 103, 104]})
print(df)
```

```python
# set_index("列名") 将指定列设置为索引
# inplace = True 表示在原对象上修改
print(df.set_index("id",inplace=True))
```

```python
df.reset_index(inplace=True) # 重置索引 重新设置为最初的默认索引 0 1 2 3
print(df)
```

```python
# 修改列名
df = pd.DataFrame(
    {"age": [20, 30, 40, 10],
     "name": ["张三", "李四", "王五", "赵六"],
     "id": [101, 102, 103, 104]})
df.set_index("id",inplace=True)
print(df)
```

```python
# 通过rename函数修改
df.rename(
    index = {101 : "一" , 102 : "二" , 103 : "三",104 : "四"},
    columns= {"age" : "年龄","name" : "姓名"},
    inplace = True
)
print(df)
```

```python
print(df.index) # 获取属性 并且打印
print(df.columns)
```

```python
# 通过属性修改
df.index = ["001" , "002" , "003" , "004"]
df.columns = ["new_age","new_name"]
print(df)
```

```python
# 新增一列 新增的数据对应行数即可
df["phone"] = ["135 xxxx xxxx","135 xxxx xxxx","135 xxxx xxxx","135 xxxx xxxx"]
print(df)
```

```python
# drop() 删除指定的行或者列
df.drop("phone",axis=1,inplace=True)
print(df)
```

```python
# 通过del关键字删除
del df["new_age"] # 删除年龄列
print(df)
```

```python
# 插入某一列数据
df.insert(loc=0,column="phone",value=["135 xxxx xxxx","135 xxxx xxxx","135 xxxx xxxx","135 xxxx xxxx"])
print(df)
```

```python
df.insert(loc=0,column="phone",value=df["new_age"] * df.index) # 存在数值类型年龄列 是可以这样书写的 
print(df)
```
