# DataFrame导入导出操作

```python
from os.path import sep

# 将DataFrame保存到csv文件中

import pandas as pd
import os

df = pd.DataFrame({"age" : [10,20,30],"name" : ["赵四","小宝","刘能"],"id" : [101,102,103]})
print(df)
```

```python
os.makedirs("data",exist_ok=True) # 表示如果此文件夹存在不会报错

df.to_csv("data/df1.csv")
```

```python
# 导出为tsv格式的
df.to_csv("data/df2.tsv",sep="\t")
```

```python
df.to_pickle("data/df.pkl")
df.to_excel("data/df.xlsx")

# df_dict = df.to_dict()
# df.to_hdf("data/df.h5", key="df")
# df.to_html("data/df.html")
# df.to_json("data/df.json")
# df.to_feather("data/df.feather")
```

```python
df.to_clipboard() # 导出到剪切板
```
