# 将数据保存到数据库中

```python
import pandas as pd
from  sqlalchemy import create_engine

df = pd.DataFrame({"age" : [10,20,30],"name" : ["赵四","小宝","刘能"],"id" : [101,102,103]})

user = "root"
password = "9999"
port = "3306"
host = "127.0.0.1" # localhost
db_name = "ai0331"

engine = create_engine(
    f"mysql+pymysql://{user}:{password}@{host}:{port}/{db_name}?charset=utf8mb4"
)

# name 表示创建的表格名称
# con 使用哪个引擎操作数据库
# if_exists 如果存在则覆盖表
# index = False 不保存索引
df.to_sql(name = "test_table",con=engine,if_exists="replace",index=False)
```

```python
df = pd.read_sql("test_table",engine)
df
```
