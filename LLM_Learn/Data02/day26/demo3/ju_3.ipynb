# DataFrame 常用方法

```python
import pandas as pd

df = pd.DataFrame(
    data = {
        "id" : [101,102,103,104,105,106,101],
        "name" : ["赵四","小宝","富贵","大拿","刘能","张三","赵四"],
        "age" : [10,20,30,40,50,None,10]
    },
    index = ["aa","bb","cc","dd","ee","ff","aa"]
)
df
```

```python
# head() 默认查看前5行数据
print(df.head())
```

```python
print(df.head(3)) # 指定参数查看前3行数据
```

```python
# tail() 默认查看后5行数据
print(df.tail())
```

```python
# tail(2) 指定参数查看后2行数据
print(df.tail(2))
```

```python
# isin() 每个单元格都与集合中的数据进行匹配 查看是否存在
print(df.isin([103, 106]))
```

```python
# isna() 判断是否为缺失值 每一个单元格都匹配
df = pd.DataFrame(
    data = {
        "id" : [101,102,103,104,105,106,101],
        "name" : ["赵四","小宝","富贵","大拿","刘能","张三","赵四"],
        "age" : [10,20,30,40,50,None,10]
    },
    index = ["aa","bb","cc","dd","ee","ff","aa"]
)
print(df.isna())
```

```python
# sum() 求和 因为DataFrame是一个二维表结构 所以必须指定 要根据那一列求和

df["age"]

```

```python
# 将age列求和计算
print(df["age"].sum())
```

```python
# 将age列求平均值
print(df["age"].mean())
```

```python
# 将age列求最大值
print(df["age"].max())
```

```python
# 将age列求最小值
print(df["age"].min())
```

```python
# 统计年龄方差
print(df["age"].var())
```

```python
# 统计年标准差
print(df["age"].std())
```

```python
# 统计年龄中位数
print(df["age"].median())
```

```python
# 统计年龄众数
print(df["age"].mode())
```

```python
# 分位数
print(df["age"].quantile(0.5))
```

```python
# 展示当前DataFrame的基本信息
df.info()
```

```python
# 查看常见的统计指标
# 只会统计数值列 非数值列
print(df.describe())
```

```python
df = pd.DataFrame(
    data = {
        "id" : [101,102,103,104,105,106,101],
        "name" : ["赵四","小宝","富贵","大拿","刘能","张三","赵四"],
        "age" : [10,20,30,40,50,None,10]
    },
    index = ["aa","bb","cc","dd","ee","ff","aa"]
)
df
```

```python
# value_counts() 统计每一行数据出现的次数 (整行重复才算  )
print(df.value_counts())
```

```python
# count() 统计数量 统计每一列 非空数据的数量
print(df.count())
```

```python
# drop_duplicates() 去除重复的的元素 只保留第一次出现的
print(df.drop_duplicates())
```

```python
# duplicated() 标记某一行是否是重复行的 (一行的所有列都重复才算 )
print(df.duplicated())
```

```python
# duplicated()  可以指定列 来单独查看某一列哪些重复内容
df = pd.DataFrame(
    data = {
        "id" : [101,102,103,104,105,106,101],
        "name" : ["赵四","小宝","富贵","大拿","刘能","张三","赵四"],
        "age" : [10,10,30,30,50,None,10]
    },
    index = ["aa","bb","cc","dd","ee","ff","aa"]
)

print(df.duplicated(subset="age"))
```

```python
# 随机采样 sample() 默认采集一个
print(df.sample())
```

```python
# replace() 替换 全表替换
print(df.replace(10, 100))
```

```python
# equals() 比较两个DataFrame 数据 以及 索引 是否完全一致
df1 = pd.DataFrame(data = {"id" : [101,102,103], "name" : ["赵四","广坤","小宝"],"age" : [10,20,30]})
df2 = pd.DataFrame(data = {"id" : [101,102,103], "name" : ["赵四","广坤","小宝"],"age" : [10,20,30]})

print(df1.equals(df2))

```

```python
# 累计最大值
df = pd.DataFrame({"A" : [2,5,3,7,4],"B" : [1,6,2,8,3]})
print(df)
```

```python
# 累计最大值 cummax() 纵向
print(df.cummax())
```

```python
# 累计最大值 cummax() 纵向
print(df.cummax(axis=0))
```

```python
# 横向
print(df.cummax(axis=1))
```

```python
print(df.cummax(axis="index")) # 纵向
```

```python
print(df.cummax(axis="columns")) # 横向
```

```python
df = pd.DataFrame({"A" : [2,5,3,7,4],"B" : [1,6,2,8,3]})
print(df.cummin())
```

```python
print(df.cummin(axis=0)) # 纵向
```

```python
print(df.cummin(axis=1)) # 横向
```

```python
# 累计求和  cumsum()
print(df.cumsum())
```

```python
print(df.cumsum(axis=0))
```

```python
print(df.cumsum(axis=1))
```
