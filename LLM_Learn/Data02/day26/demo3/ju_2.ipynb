# DataFrame 的创建

```python
import pandas as pd
# 根据字典创建 指定列和数据
df = pd.DataFrame({
    "id" : [1,2,3,4],
    "name" : ["赵四","广坤","小宝","刘能"],
    "age" : [25,20,30,35]
})
df
```

```python
# 根据参数 来指定数据 列 以及 索引
df = pd.DataFrame(
    data={"age" : [22,23,34,45],"name":["赵四","广坤","小宝","刘能"]},
    columns= ["age","name"],
    index = [101,102,103,104]
)
df
```

```python
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] },
    index = ["aa","bb","cc","dd"]
)
df
```

# DataFrame属性

```python
# 行索引
print(df.index)
```

```python
# 列索引
print(df.columns)
```

```python
# 将DataFrame转换为二维的数据
print(df.values)
print(type(df.values))
```

```python
# 维度 ndim
print(df.ndim)
```

```python
# 形状
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] },
    index = ["aa","bb","cc","dd"]
)
print(df.shape)
```

```python
#元素总个数 行数 * 列数  所有单元格的数量
print(df.size)
```

```python
# 每一列的数据类型
print(df.dtypes)
print(type(df.dtypes)) # 最终的结果还是Series类型的
```

```python
# 行列转置

print(df.T)
```

```python
# loc[] 根据标签 获取单个(某一行)或者是切片(多行)操作
df = pd.DataFrame(
    data = {"id" : [101,102,103,104],"name" :["赵四","广坤","小宝","刘能"],"age" :[22,23,34,45] },
    index = ["aa","bb","cc","dd"]
)

# 获取标签为'aa'的这一行数据
df.loc["aa"]
```

```python
df
```

```python
# 切片操作 包含起始和结束

df.loc["aa" : "bb"]
```

```python
# 获取所有数据 并且只展示 id 和 name 这两列
df.loc[:,["id","name"]]
```

```python
# iloc[] 都是根据数值下标 隐式索引操作的
# 获取第一行数据
df.iloc[0]
```

```python
# 切片获取指定行数据
df.iloc[0:3]
```

```python
# 切片操作，只要第0(一)列数据
df.iloc[0:3,0]
```

```python
# at 按照标签索引获取单个数据
# 只能获取单个数据 获取aa这一行的名字数据
df.at["aa","name"]
```

```python
# 只能获取单个数据 获取aa这一行的id数据
df.at["aa","id"]
```

```python
# 只能获取单个数据 获取aa这一行的年龄数据
df.at["aa","age"]
```

```python
# iat 根据数值索引获取数据 也是只能获取单个的 数据
# 行索引 和 列索引 都是从0开始的
df.iat[0,1]
```
