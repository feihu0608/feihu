# Series对象的数学计算

```python
import pandas as pd

s1 = pd.Series({"a" : 1, "b" : 2, "c" : 3})
print(s1 * 10)
```

```python
# 相同位置的元素进行计算
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([1,2,3,4])
print(s1 + s2)
print(s1 - s2)
print(s1 * s2)
print(s1 / s2)
```

```python
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([1,2,3,4],index=[1,2,3,4])
print(s1 + s2)
```
