 # 睡眠质量分析

```python
import pandas as pd

df = pd.read_csv("sleep.csv")
print(df)
```

```python
# 统计不同睡眠时长 不同压力等级下的平均睡眠质量

print(df.groupby(["sleep_duration", "stress_level"])["sleep_quality"].mean().head(50))
```

```python
#  对睡眠时间 进行连续的分箱操作
sleep_cut = pd.cut(df["sleep_duration"], [0,5,6,7,8,9,10,11,12])


# 对压力等级分箱操作

stress_cut = pd.cut(df["stress_level"], 4)

print(pd.pivot_table(
    df,
    values="sleep_quality",
    index=[sleep_cut, stress_cut],
    aggfunc="mean"
))




```

```python
# 多维度透视表
# 不同睡眠时长 不同职业 不同压力等级的睡眠质量

print(pd.pivot_table(
    df,
    values="sleep_quality",
    index=[sleep_cut, stress_cut,"occupation"],
    aggfunc="mean"
))
```

```python
pd.pivot_table(
    df,
    values="sleep_quality",
    index=[sleep_cut, stress_cut],
    columns=["occupation"],
    aggfunc="mean"
)
```

```python
# 多维度透视表
# 不同睡眠时长 不同职业 不同压力等级  不同性别  的睡眠质量
print(pd.pivot_table(
    df,
    values="sleep_quality",
    index=[sleep_cut, stress_cut,"occupation","gender"],
    aggfunc="mean"
))
```

```python
print(pd.pivot_table(
    df,
    values="sleep_quality",
    index=[sleep_cut, stress_cut],
    columns=["occupation","gender"],
    aggfunc="mean"
))
```
