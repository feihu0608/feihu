# transform函数 数据转换操作

```python
import pandas as pd

df = pd.read_csv("employees.csv")

print(df)




```

```python
# 统计每个部门每个人薪资与部门薪资的差值

print(df.groupby("department_id")["salary"].transform(lambda x: x - x.mean()))
```

```python
# 通过transform()按分组使用平均值填充缺失值

s1 = pd.Series(df.index.tolist()).sample(10)

print(s1)
```

```python
# 将对应的数据 薪资列 置为NA
df.loc[s1,"salary"] = pd.NA
print("********填充缺失值之前********")
print(df.groupby("department_id")["salary"].agg(["size", "count"]))
```

```python
import numpy as np


# 使用各个部门的平均薪资 填充缺失值

def input_mean(group_data):
    # 如果这个部门的平均值也是缺失值 则使用0填充
    if np.isnan(group_data.mean()):
        return 0
    # 否则填充平均值
    return group_data.fillna(group_data.mean())


df["salary"] = df.groupby("department_id")["salary"].transform(input_mean)

print(df.groupby("department_id")["salary"].agg(["size", "count"]))
```

# filter函数

```python
# 属于分组之后过滤的操作  类似于mysql中的having
# 按照部门分组 只保留部门人数大于40的信息

print(df.groupby("department_id").filter(lambda x: len(x) > 40))
```
