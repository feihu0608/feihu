# 时间索引

```python
# DatetimeIndex
# to_datetime() 如果传入的是一个时间序列  那么返回值就是DatetimeIndex

import pandas as pd
# format="mixed" 表示可以识别不同格式的日期
print(pd.to_datetime(["4th of July, 2015", "2015-Jul-6", "07-07-2015", "20150708"], format="mixed"))
```

```python
df = pd.read_csv("weather.csv")
print(df.info())
```

```python
df = pd.read_csv("weather.csv")
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)
df.info()
```

```python
# 切片操作
# 筛选出 2013年1月 到 2013年6月的天气数据

print(df.loc["2013-01":"2013-06"])
```

```python
# 获取某一年的数据
print(df.loc["2015"])
```

```python
# TimedeltaIndex 时间差索引
df = pd.read_csv("weather.csv", parse_dates=[0])
df_date = df["date"]

df["timedelta"] = df_date - df_date[0]

df.set_index("timedelta", inplace=True)

print(df.info())
```
