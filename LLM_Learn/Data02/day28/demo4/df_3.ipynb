# 统计异常值

```python
# IQR 统计方式

import pandas as pd

df = pd.DataFrame({"price":[100,101,102,103,104,105,10000]})
print(df)
```

```python
# 找到四分位数 Q1 0.25分位

Q1 = df["price"].quantile(0.25)

# 找到哦四分位数 Q3 0.75 分位
Q3 = df["price"].quantile(0.75)

# 四分位距  IQR  Q3 -Q1  得到中间50% 数据波动范围

IQR = Q3 - Q1

# 计算正常范围的下限
lower_bound = Q1 - 1.5 * IQR

# 计算正常范围的上限
upper_bound = Q3 + 1.5 * IQR

print(f"正常范围{lower_bound}~{upper_bound}")

print(df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)])

```
