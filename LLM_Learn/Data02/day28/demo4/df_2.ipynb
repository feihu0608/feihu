# 生成时间序列

```python
import pandas as pd
# 指定了起始和结束 生成序列
print(pd.date_range("2022-10-10", "2022-10-20"))
```

```python
# 只指定起始
print(pd.date_range("2022-12-12", periods=5))
```

```python
# 指定起始时间 以及 指定频率

print(pd.date_range("2022-12-12", periods=5, freq="h"))
```

```python
# 生成季度时间序列
# 默认情况 3 6 9 12
# 季度频率 ： 1 4 7 10
print(pd.date_range("2015-05-03", periods=10, freq="QE-JAN"))

```

```python
# 指定具体到周几的频率

print(pd.date_range("2026-05-13", periods=10, freq="W-FRI"))
```

```python
# 指定小时和分钟 生成时间序列
print(pd.date_range("2026-05-13", periods=10, freq="2h30min"))
```

```python
# 统计最高温度 最低温度的平均值

df = pd.read_csv("weather.csv",parse_dates=["date"])

df.set_index("date", inplace=True)
print(df[["temp_max", "temp_min"]].resample("YE").mean())
```

```python
import numpy as np
import matplotlib.pyplot as plt  # 导入matplotlib

x = np.linspace(0, 10, 100)  # 创建x轴的数据
y1 = np.sin(x)  # 创建y轴的数据
y2 = np.cos(x)  # 创建y轴的数据

fig, ax = plt.subplots(2, figsize=(10, 6))  # 创建画布，并指定画布大小

ax[0].set_xlim(0, 10)  # 设置x轴的范围
ax[0].set_ylim(-1, 1)  # 设置y轴的范围
ax[0].set_xlabel("x")  # 设置x轴的标签
ax[0].set_ylabel("sin(x)")  # 设置y轴的标签
ax[0].set_title("sin")  # 设置子图的标题
ax[0].plot(x, y1)  # 绘制曲线

ax[1].plot(x, y2)  # 绘制曲线
ax[1].set_xlim(0, 10)  # 设置x轴的范围
ax[1].set_ylim(-1, 1)  # 设置y轴的范围
ax[1].set_xlabel("x")  # 设置x轴的标签
ax[1].set_ylabel("cos(x)")  # 设置y轴的标签
ax[1].set_title("cos")  # 设置子图的标题

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams["font.sans-serif"] = ["SimHei"]  # 指定中文字体
rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

def year_color(x):
    """添加一列，为不同年份的数据添加不同的颜色"""
    match x.year:
        case 2012:
            return "r"
        case 2013:
            return "g"
        case 2014:
            return "b"
        case 2015:
            return "k"

df = pd.read_csv("weather.csv")
df["date"] = pd.to_datetime(df["date"])
df["color"] = df["date"].apply(year_color)
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
# 绘制散点图，横轴为最高气温，纵轴为降水量
# c设置颜色,alpha设置透明度
ax1.scatter(df["temp_max"], df["precipitation"], c=df["color"], alpha=0.5)
ax1.set_xlabel("最高气温")
ax1.set_ylabel("降水量")
plt.show()
```
