# 分箱函数 cut

```python
import pandas as pd

df = pd.read_csv("employees.csv")
print(df)
```

```python
# 针对整个表所有行 进行分箱
print(pd.cut(df["salary"], bins=3))
```

```python
# 针对于指定的数据 分箱操作
print(pd.cut(df.iloc[9:16]["salary"], bins=5))
```

```python
# 手动指定范围 进行分箱操作
# 分成两个区间 0~ 10000  10000 ~ 20000
print(pd.cut(df["salary"], [0, 10000, 20000]))
```

```python
# 指定标签 分箱
# 将薪资分成三个区间 然后分别以  lower medium high
print(pd.cut(df["salary"], bins=3, labels=["low", "mid", "high"]))
```
