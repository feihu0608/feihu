# agg函数

```python
import pandas as pd

df = pd.read_csv("employees.csv")
print(df)
```

```python
group_obj = df.groupby("department_id")["salary"]
print(group_obj)
```

```python
# 可以重复利用以上的分组对象 执行聚合操作

print(group_obj.mean())
```

```python
print(group_obj.min())
```

```python
print(group_obj.max())
```

```python
print(group_obj.sum())
```

```python
# agg函数 原名：aggregate()
# 一次性执行多种不同的聚合操作

print(df.groupby("department_id")["salary"].agg(["max", "min", "sum", "mean"]))
```

```python
# agg函数还支持传入字典 可以指定不同的列 执行不同的聚合操作
print(df.groupby("department_id").agg(
    {
        "job_id": "nunique",
        "commission_pct": "mean"
    }
))
```

```python
# 因为执行完聚合操作之后  依然使用之前的列名来呈现内容  语义不准确 所以我们可以重命名列名
print(df.groupby("department_id").agg(
    {
        "job_id": "nunique",
        "commission_pct": "mean"
    }
).rename(columns={"job_id": "工种数量","commission_pct" : "佣金比例平均值"}))
```

```python
# 如果我们需要针对分组之后的数据做一些个性化(即PandasAPI中未提供的)的操作
# 我们可以自定义函数实现

# 统计每个部门员工姓氏首字母包含了哪些

def get_first(group_data):
    result = set() # 创建空集合 用于存放不同的形式首字母
    for name in group_data:
        result.add(name[0])
    return result



print(df.groupby("department_id")["last_name"].agg(get_first))
```
