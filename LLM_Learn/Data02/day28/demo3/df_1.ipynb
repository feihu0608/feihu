# 日期时间类型和索引

```python
from datetime import datetime

import pandas as pd

d1 = datetime(year=2018, month=6, day=23)
d2 = datetime.now()

# 1S = 1000ms  1ms = 1000us  1us = 1000ns  1秒=10亿纳秒
print(d1)
print(d2)

# d2.weekday() 0 = 周一 1=周二
print(d2.year, d2.month, d2.day,d2.weekday())
print(d2.strftime("%A")) # 英文全拼
print(d2.strftime("%a")) # 缩写

print(d2 - d1)
print(type(d2 - d1))
```

```python
df = pd.read_csv("weather.csv")
# 在我们使用to_datetime函数操作DataFrame中的某一列 会将日期自动优化为datetime64类型
df["ymd"] = pd.to_datetime(df["date"])
print(df.info())
```

```python
# 按照天周期进行划分

df["day"] = df["ymd"].dt.to_period("D")
print(df.info())
```

```python

date1 = pd.to_datetime("2022-12-12")
print(date1, type(date1)) # 直接使用 不涉及DataFrame 不会优化 默认类型还是 Timestamp
```
