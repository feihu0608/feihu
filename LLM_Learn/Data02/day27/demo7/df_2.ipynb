# 剔除缺失值

```python
import pandas as pd

s1 = pd.Series([1,pd.NA,None])
# 删除缺失值
print(s1.dropna())
```

```python
df = pd.DataFrame([[1, pd.NA, 2], [2, 3, 5], [pd.NA, 4, 6]])
print(df)
```

```python
print(df.dropna()) # 针对于DataFrame 如果当前行有缺失值 则删除整行
```

```python
print(df.dropna(axis=1)) # 按照列删除 如果某一列有缺失值 则删除整列
```

```python
df = pd.DataFrame([[1, pd.NA, 2], [pd.NA, pd.NA, 5], [pd.NA, pd.NA, pd.NA]])
print(df)
```

```python
print(df.dropna(how="all"))
```

```python
print(df.dropna(thresh=2)) #至少有两个非缺失值 保留数据 否则 即删除
```

```python
print(df.dropna(subset=[0])) # 只看第0列 如果第0列为缺失值 则删除整行
```
