# 缺失值的处理

```python
import pandas as pd
import numpy as np
s = pd.Series([np.nan, None, pd.NA])

# isnull()函数 以上三种都认为是缺失值
print(s.isnull())
```

```python
# 不把空字符串识别为缺失值
df = pd.read_csv("weather_withna.csv",keep_default_na=False)
print(df)
```

```python
# 在读取文件的时候 将某一个值 设置为缺失值
df = pd.read_csv("weather_withna.csv",na_values=["2015-12-31"])
print(df)
```

```python
# 统计缺失值的数量
print(df.isnull().sum())
```
