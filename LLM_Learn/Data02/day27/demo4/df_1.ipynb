# 员工信息数据分析

```python
import pandas as pd

df = pd.read_csv("employees.csv")
print(df)
```

```python
# 查看前5行数据
print(df.head())
```

```python
# 查看DataFrame的基本信息
df.info()
```

```python
# 查看常用的统计指标
print(df.describe())
```

```python
# 所有字段名字
print(df.columns)
```

```python
# 找出薪资最高 或者 最低员工的信息

print(df[df["salary"] == df["salary"].min()])
```

```python
# 最高薪资
print(df[df["salary"] == df["salary"].max()])
```

```python
# 通过loc的方式实现相同的效果
print(df.loc[df["salary"] == df["salary"].max()])
```

```python
# 按照排序的思路实现
print(df.sort_values("salary",ascending=False).head(1)) # 最高薪资
print(df.sort_values("salary").head(1)) # 最低薪资
```

```python
# 查出薪资最高的前10位员工

print(df.nlargest(10, "salary"))
```

```python
# 查看有哪些部门编号
print(df["department_id"].unique())
```

```python
# 查看每个部门的员工人数
print(df.groupby("department_id")["employee_id"].count())
```

```python
# 找出平均薪资最高的部门编号
print(df.groupby("department_id")["salary"].mean().nlargest(1))
```

```python
df.groupby("department_id")["employee_id"].count().rename("employee_count").plot(kind="bar")
```
