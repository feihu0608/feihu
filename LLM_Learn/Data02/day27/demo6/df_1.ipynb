# 数据合并操作

```python
import pandas as pd
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
df2 = pd.DataFrame(data={"b": [7, 8], "c": [10, 11]}, index=[2, 3])

# 没有对应的列 使用NaN填充
print(pd.concat([df1, df2]))
```

```python
# 取交集 只获取两个DataFrame共同的列
print(pd.concat([df1, df2],join="inner"))
```

```python
# merge() 合并操作
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame({"employee": ["Lisa", "Bob", "Jake", "Sue"], "hire_date": [2004, 2008, 2012, 2014]})
print(df1)
```

```python
print(df2)
```

```python
# 使用merge()合并操作 会自动匹配相同的列 作为关联依据
df3 = pd.merge(df1,df2)  # ,on="employee" 参数可写可不写
print(df3)
```

```python
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame({"group": ["Accounting", "Engineering", "HR"], "supervisor": ["Carly", "Guido", "Steve"]})
print(df1)
```

```python
print(df2)
```

```python
print(pd.merge(df1, df2))
```

```python
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame(
    {
        "group": ["Accounting", "Accounting", "Engineering", "Engineering", "HR", "HR"],
        "skills": ["math", "spreadsheets", "coding", "linux", "spreadsheets", "organization"],
    }
)
print(df1)
```

```python
print(df2)
```

```python
print(pd.merge(df1, df2)) # 对于多对多 最终的结果行数就比较多了  笛卡尔积
```

```python
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame({"employee": ["Lisa", "Bob", "Jake", "Sue"], "hire_date": [2004, 2008, 2012, 2014]})
print(df1)
```

```python
print(df2)
```

```python
print(pd.merge(df1, df2,on="employee"))  # 显式指定需要使用哪一列进行关联
```

```python
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame({"name": ["Bob", "Jake", "Lisa", "Sue"], "salary": [70000, 80000, 120000, 90000]})
print(df1)
```

```python
print(df2)
```

```python
# 如果没有相同的列 必须手动指定使用哪些列进行关联
print(pd.merge(df1, df2,left_on="employee", right_on="name"))
```

```python
df1 = pd.DataFrame(
    {"employee": ["Bob", "Jake", "Lisa", "Sue"], "group": ["Accounting", "Engineering", "Engineering", "HR"]}
)
df2 = pd.DataFrame({"employee": ["Bob", "Jake", "Lisa", "Sue"], "salary": [70000, 80000, 120000, 90000]})


df1.set_index("employee", inplace=True)
df2.set_index("employee", inplace=True)

# 使用索引关联
print(pd.merge(df1, df2, left_index=True, right_index=True))

```
