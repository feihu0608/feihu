# 数据合并操作

```python
import pandas as pd

df1 = pd.DataFrame({"id":[1,2],"name":["赵四","广坤"]})
df2 = pd.DataFrame({"id":[1,2],"age":[20,30]})


print(df1)

```

```python
print(df2)
```

```python
# 使用join函数合并数据 横向 增加多列
# 将df2的索引设置为id
# on 指定关联列
print(df1.join(df2.set_index("id"),on="id"))
```

```python
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'value1': [1, 2, 3]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'value2': [4, 5, 6]
})

# 直接关联会报错 因为两个表都有相同的列
# 所以我们通过取新的列名的方式 避免这种冲突   
print(df1.join(df2,lsuffix="_left",rsuffix="_right"))
```
