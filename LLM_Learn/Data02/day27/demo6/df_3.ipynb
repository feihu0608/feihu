# 数据合并操作  merge

```python

import pandas as pd

df1 = pd.DataFrame({"name": ["Peter", "Paul", "Mary"], "food": ["fish", "beans", "bread"]}, columns=["name", "food"])
df2 = pd.DataFrame({"name": ["Mary", "Joseph"], "drink": ["wine", "beer"]}, columns=["name", "drink"])
print(df1)
```

```python
print(df2)
```

```python
print(pd.merge(df1, df2)) # ,how="inner" 默认效果 取交集
```

```python
print(pd.merge(df1, df2,how="outer")) # 类似于MySQL中的union操作 合并结果集
```

```python
df1 = pd.DataFrame({"name": ["Peter", "Paul", "Mary"], "food": ["fish", "beans", "bread"]}, columns=["name", "food"])
df2 = pd.DataFrame({"name": ["Mary", "Joseph"], "drink": ["wine", "beer"]}, columns=["name", "drink"])
print(df1)
```

```python
print(pd.merge(df1, df2,how="left")) # 类似于MySQL中的左连接
```

```python
print(df2)
```

```python
print(pd.merge(df1, df2,how="right")) # 类似于MySQL中的右连接
```

```python
df1 = pd.DataFrame({"name": ["Bob", "Jake", "Lisa", "Sue"], "rank": [1, 2, 3, 4]})
df2 = pd.DataFrame({"name": ["Bob", "Jake", "Lisa", "Sue"], "rank": [3, 1, 4, 2]})

# 使用merge()合并 如果有相同列 会自动取别名
print(pd.merge(df1, df2,on="name")) # 因为这两列名字都一样 所以必须指定on参数
```

```python
print(pd.merge(df1, df2,on="name",suffixes=("_df1","_df2"))) # 因为这两列名字都一样 所以必须指定on参数
```
