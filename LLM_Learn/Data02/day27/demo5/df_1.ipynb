# Pandas 合并操作

```python
import pandas as pd
s1 = pd.Series(["A", "B"], index=[1, 2])
s2 = pd.Series(["D", "E"], index=[4, 5])
s3 = pd.Series(["G", "H"], index=[7, 8])

# 默认为上下堆叠 拼接成一个新的Series对象
# Series拼接
print(type(pd.concat([s1, s2, s3])))
pd.concat([s1, s2, s3])
```

```python
# Series拼接
print(pd.concat([s1, s2, s3], axis=1))
```

```python
# DataFrame与Series拼接
df1 = pd.DataFrame(data= {"a" : [1,2], "b" : [3,4]},index=[1,2])
print(df1)
```

```python
s1 = pd.Series(data=[7,10],index=[1,2],name="a")
print(s1)
```

```python
# 按行连接  上下堆叠 axis = 0
# 列名对齐 有就填充 没有就NaN
print(pd.concat([df1, s1]))
```

```python
print(pd.concat([df1, s1], axis=1))
```

```python
# DataFrame与DataFrame拼接
df1 = pd.DataFrame(data={"a": [1, 2], "b": [4, 5]}, index=[1, 2])
df2 = pd.DataFrame(data={"a": [7, 8], "b": [10, 11]}, index=[1, 2])
print(pd.concat([df1, df2])) # 默认上下堆叠
```

```python
print(pd.concat([df1, df2],axis=1))
```

```python
print(pd.concat([df1, df2],ignore_index=True))
```
