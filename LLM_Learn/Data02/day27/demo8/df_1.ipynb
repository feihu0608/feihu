# 填充缺失值

```python
import pandas as pd


df = pd.read_csv("weather_withna.csv")
print(df)
```

```python
# 填充缺失值函数 fillna(字典)
print(df.fillna({"temp_max": 25, "temp_min": -25}).tail())
```

```python
# 将缺失值 填充为这一列的平均值
# 获取指定的列 计算平均值 再使用fillna进行填充  会根据我们指定的列 自动匹配进行填充
print(df.fillna(df[["precipitation", "temp_max", "temp_min", "wind"]].mean()))

```

```python
import numpy as np

# ffill forward 向前  用前边的非缺失值填充当前缺失值
# bfill backward 向后   用后边的非缺失值填充当前缺失值

df = pd.DataFrame([None,10,np.nan,None,30,None,np.nan])

print(df)
```

```python
df.ffill(inplace=True) # 原地修改
df.bfill(inplace=True) # 原地修改
print(df)
```

```python
df = pd.read_csv("weather_withna.csv")
df.ffill(inplace=True)
df.bfill(inplace=True)
print(df)
```
