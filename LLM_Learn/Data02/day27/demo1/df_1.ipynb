# DataFrame 日期处理

```python
import pandas as pd

df = pd.DataFrame({
    "gmv" : [100,200,300,400],
    "trade_date" : ["2025-01-06","2023-10-31","2023-12-31","2023-01-05"]
})

print(df)
```

```python
# to_datetime(列) 将普通列(str)类型转换为日期时间类型
print(df["trade_date"].dtype) # str 类型
print(pd.to_datetime(df["trade_date"])) # datetime64
```

```python
df["ymd"] = pd.to_datetime(df["trade_date"])
print(df)
```

```python
# 根据指定的datetime类型的列 获取这一列对应的年月日信息

print(df["ymd"].dt.year)
print(df["ymd"].dt.month)
print(df["ymd"].dt.day)
```

```python
# 将获取到的年月日再次添加到DataFrame对象中

df["yy"],df["mm"],df["dd"] = df["ymd"].dt.year, df["ymd"].dt.month, df["ymd"].dt.day

print(df)
```

```python
# 获取周几 星期几
df["week"] = df["ymd"].dt.day_name()
print(df)
```

```python
# 获取当前日期属于哪个季度

df["quarter"] = df["ymd"].dt.quarter
print(df)
```

```python
# 判断是否是月底
df["is_month_end"] = df["ymd"].dt.is_month_end
print(df)
```

```python
# 判断是否为年底 一年的最后一天
df["is_year_end"] = df["ymd"].dt.is_year_end
print(df)
```

```python
# to_period() 函数 用户处理日期周期

df["ystat"] = df["ymd"].dt.to_period("Y")
print(df)
```

```python
# 按照月份周期统计
df["mstat"] = df["ymd"].dt.to_period("M")
print(df)
```

```python
# 按照  周  周期统计
df["wstat"] = df["ymd"].dt.to_period("W")
print(df)
```

```python
# 按照  季度  周期统计
df["qstat"] = df["ymd"].dt.to_period("Q")
print(df)
```
