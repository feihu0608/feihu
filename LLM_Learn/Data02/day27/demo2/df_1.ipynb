# 天气数据分析

```python
import pandas as pd

df = pd.read_csv("weather.csv")
print(df)
```

```python
# 查看DataFrame属性
print(type(df))
```

```python
# 形状
print(df.shape)
```

```python
# 列信息
print(df.columns)
```

```python
print(df.info())
```

```python
# 获取某一列 查看类型
# 获取某一列 并且 如下书写格式 获取到的是一个Series
s1 = df["date"]
print(type(s1))
```

```python
#获取多列 得到的是一个DataFrame类型
d1 = df[["date","temp_max"]]
print(type(d1))
```

```python
d2 = df[["date"]]
print(type(d2))
```

```python
# 访问 iloc[] 隐式索引的方式  自动生成的索引
print(df.iloc[0])
```

```python
print(df.iloc[1])
```

```python
# 没有通过 set_index() 显示设置索引 那么 loc依然可以使用隐式索引进行访问
print(df.loc[0])
```

```python
# 没有通过 set_index() 显示设置索引 那么 loc依然可以使用隐式索引进行访问
print(df.loc[1])
```

```python
print(df.iloc[-1]) # 访问最后一行
```

```python
# 或去某一行的某一列数据
# 因为要根据标签 获取某一列的数值 所以 必须使用loc
print(df.loc[1, "precipitation"])
```

```python
# 获取所有行的某一列数据

print(df.loc[:, "precipitation"])
```

```python
# 按照数值进行切片 以及 获取 数值对应的列
# 只要涉及到数值操作的 iloc
print(df.iloc[:, [1, 3, -1]])
```

```python
# 按照数值进行切片 列的范围 切片
print(df.iloc[:10, 2:6])
```
