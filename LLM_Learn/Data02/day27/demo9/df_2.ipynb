# Pandas 分组操作

```python
import pandas as pd
df = pd.read_csv("employees.csv")
print(df)
```

```python
grou_obj = df.groupby("department_id")
print(grou_obj)
```

```python
print(grou_obj.groups) # 查看分组信息
```

```python
# 获取某一个分组的完整数据

print(df.groupby("department_id").get_group(50))
```

```python
# 计算每个部门的平均薪资

print(df.groupby("department_id")["salary"].mean())
```

```python
# groupby对象也属于可迭代对象
# dep_id 组名 部门名称
# 该组的完整数据
for dep_id , group in df.groupby("department_id"):
    print(f"当前组{dep_id},数据行数，列数{group.shape}")
    print(group.iloc[:,0:3])
    print("*" * 50)

```

```python
# 针对员工表的数据 统计每个部门  每个工种   平均薪资 提成比例 计算平均值

emp_mean_df = df.groupby(["department_id", "job_id"])[["salary", "commission_pct"]].mean()
print(emp_mean_df)
```

```python
# 分组之后的行索引

print(emp_mean_df.index)
print(type(emp_mean_df.index)) # 复合索引
```
