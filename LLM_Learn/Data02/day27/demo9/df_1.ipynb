# apply 函数

```python
# apply()函数可以对DataFrame或Series的数据进行逐行、逐列或逐元素的操作。可以使用自定义函数对数据进行变换、计算或处理，通常用于处理复杂的变换逻辑，或者处理不能通过向量化操作轻松完成的任务。
```

```python
import pandas as pd

s1 = pd.Series([1,2,3,4])

def func(item):
    return item * 2


print(s1.apply(func))
```

```python
print(s1.apply(lambda item: item * 2))
```

```python
s1 = pd.Series([1,2,3,4])

def func(item,p1):
    return item * p1

# 必须使用关键字参数传参
print(s1.apply(func, p1=3))
```

```python
# 编写函数 根据传入的Series(DataFrame)对象 计算总和

def func(s):
    return s.sum()


df = pd.DataFrame({"a":[1,2,3],"b":[4,5,6]})

# 根据两个Series求和计算
print(df.apply(func))


```

```python

# 根据两个Series求和计算  列对应列计算
print(df.apply(func,axis=1))
```

```python
def func(s):
    return s["a"] / s["b"]

df = pd.DataFrame({"a":[1,2,3],"b":[4,5,6]})

print(df.apply(func, axis=1))
```

```python
import numpy as np


def f(x, y):
    if y == 0:
        return np.nan
    return x / y

df = pd.DataFrame({"a": [10, 20, 30], "b": [40, 0, 60]})
print(f(df["a"], df["b"]))  # ValueError
```

```python
s1 = pd.Series([1,2,3,4,0])
print(s1 == 0)
```

```python
def f(x, y):
    if y == 0:
        return np.nan
    return x / y

df = pd.DataFrame({"a": [10, 20, 30], "b": [40, 0, 60]})
f_vec = np.vectorize(f)
print(f_vec(df["a"], df["b"]))  # [0.25  nan 0.5 ]
```

```python
@np.vectorize
def f(x, y):
    if y == 0:
        return np.nan
    return x / y

df = pd.DataFrame({"a": [10, 20, 30], "b": [40, 0, 60]})
print(f(df["a"], df["b"]))  # [0.25  nan 0.5 ]
```
