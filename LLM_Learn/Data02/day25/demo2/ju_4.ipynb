# 常用函数


```python
import numpy as np

arr1 = np.array([1,2,3,4,-5,-6])

arr1
```

```python
np.abs(arr1)
```

```python
arr2 = np.array([1.5,2.3,3.6,4,-5,-6])
# 向上取整
np.ceil(arr2)
```

```python
arr2 = np.array([1.5,2.3,3.6,4,-5,-6])

np.floor(arr2) # 向下取整
```

```python
arr2 = np.array([1.5,2.3,3.6,4,-5,-6])
np.rint(arr2) # 四舍五入
```

```python
arr2 = np.array([1.5,2.3,3.6,4,-5,-6])


np.isnan(arr2) # NaN not a number 不是一个数字为True 是的话为False

```

```python
arr2 = np.array(["a","b","c","d","e","f","g"])
np.isnan(arr2) # 类型不匹配
```

```python
arr3 = np.array([1,2,3,4])
np.multiply(arr3,2) # 乘法计算 数组中的每个元素都与指定的数据相乘
```

```python
arr3 = np.array([1,2,3,4])
np.divide(arr3,2) # 除法计算 数组中的每个元素都与指定的数据相除
```

```python
arr3 = np.array([1,2,3,4])
arr4 = np.array([1,2,3,4])
np.divide(arr3,arr4) # 对位计算
```

```python
arr3 = np.array([1,2,3,4])
arr4 = np.array([1,2,3,4])
np.multiply(arr3,arr4) # 对位计算
```

```python
arr3 = np.array([1,2,3,4])

np.where(arr3 > 2 , "ok" , "error")
```
