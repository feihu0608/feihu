# 切片操作 用法和之前一致

```python
import numpy as np

# 生成0~9序列填充数组
arr1 = np.arange(10)
arr1
```

```python
print(arr1[2]) # 使用下标访问
```

```python
print(arr1[1:3]) # 包前不包后
```

```python
print(arr1[2:]) # 从指定位置开始到最后
```

```python
print(arr1[2:9:2]) # 指定起始和结束 以及 步长
```

```python
print(arr1[slice(2,9,2)]) # 效果跟上方一致
```
