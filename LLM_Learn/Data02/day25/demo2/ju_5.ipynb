# 统计函数

```python
import numpy as np

arr1 = np.array([[1,2,3,8],[7,4,5,6]])

arr1
```

```python
print(np.mean(arr1)) # 平均值
```

```python
print(np.sum(arr1)) # 求和
```

```python
print(np.max(arr1)) # 最大值
```

```python
print(np.min(arr1)) # 最小值
```

```python
print(np.var(arr1))  # 方差
```

```python
print(np.std(arr1))  # 标准差
```

```python
print(np.argmax(arr1))  #获取最大值的下标
```

```python
print(np.argmin(arr1))  # 获取最小值下标
```

```python
arr1 = np.array([[1,2,3,8],[7,4,5,6]])
print(np.cumsum(arr1)) # 将数组展平以后 求累计和
```

```python
arr1 = np.array([[1,2,3,8],[7,4,5,6]])
print(np.cumprod(arr1)) # 将数组展平以后 求累计乘积
```
