# 创建ndarray数组方式

```python
import numpy as np

arr1 = np.zeros((2,3))
# 在Jupyter中也可以直接书写要呈现的对象名称 不需要加print 直接运行 打印内容
arr1
```

```python
arr2 = np.ones((2,3)) # 创建2行3列的 全为1的数组 类型为浮点类型 float64
arr2
```

```python
arr3 = np.empty((3,4))
arr3
```

```python
# 根据指定的数组 生成相同形状的数组
a1 = np.zeros_like(arr1)
a1
```

```python
# 根据指定的数组 生成相同形状的数组
a2 = np.ones_like(arr2)
a2
```

```python
# 根据指定的数组 生成相同形状的数组
a3 = np.empty_like(arr3)
a3
```
