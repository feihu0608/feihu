```python
import numpy as np

# 创建一个二维数组 高维度为2 表示二维数组中包含2个一维数组
# 低维度为3 表示每个一维数组中包含3个元素
arr1 = np.array([[1,2,3],[4,5,6]])   #arr1 = [2][3]

# 属性
print("维度",arr1.ndim)
print("形状",arr1.shape)
print("元素个数",arr1.size)
print("元素类型",arr1.dtype)
print("元素大小字节",arr1.itemsize)
```

```python
data = [1,2,3]

print("data列表的地址为",id(data))

a1 = np.array(data)

print("a1数组的地址为",id(a1))

a2 = np.asarray(data)

print("a2列表的地址为",id(a2))
```

```python
a3 = np.array((1,2,3,4))

print("a3的地址值是",id(a3))

# 会创建新数组
new_a1 = np.array(a3)

print("new_a1的地址值是",id(new_a1))

# 如果本身传入的就是一个ndarray asarray函数不会创建新数组 属于在原数组进行操作
new_a2 = np.asarray(a3)

print("new_a2的地址值是",id(new_a2))
```
