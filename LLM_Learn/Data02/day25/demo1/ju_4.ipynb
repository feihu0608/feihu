# 创建ndarray的方式

```python
import numpy as np

# 创建给定形状的数组 数据属于0~1之间的随机小数
a1 = np.random.rand(2,3)
a1
```

```python
# 随机整数 1 ~ 10 之间  3行2列
a2 = np.random.randint(1,10,(3,2))
a2
```

```python
# 随机小数 3 ~ 6 之间  3行3列
a3 = np.random.uniform(3,6,(3,3))
a3
```

```python
# 返回给定形状的数组，用标准正态分布(均值为0，标准差为1)的随机样本填充。
# 2行3列
a4 = np.random.randn(2,3)
a4
```
