# 基本运算和矩阵乘法

```python
import numpy as np


a1 = np.array([[1,2,3],[4,5,6]])
a2 = np.array([[4,5,6],[7,8,9]])
print(a1 + a2)# 相同位的元素进行计算
```

```python
print(a1 - a2) # 相同位的元素进行计算
```

```python
print(a1 * a2)# 相同位的元素进行计算
```

```python
print(a1 / a2)# 相同位的元素进行计算
```

```python
a1 = np.array([[1,2,3],[4,5,6]])

print(a1 + 10) # 对于标量计算 应用到元素级 即每个元素都对标量进行计算
```

```python
print(a1 - 10) # 对于标量计算 应用到元素级 即每个元素都对标量进行计算
```

```python
print(a1 * 10) # 对于标量计算 应用到元素级 即每个元素都对标量进行计算

```

```python
print(a1 / 10) # 对于标量计算 应用到元素级 即每个元素都对标量进行计算
```

```python
# 不同形状的数组进行广播计算 两个数组的维度 从右往左 每一位的维度 要么相同 要么其中一个为1
a3 = np.array([[1,2,3],[4,5,6]])  # 2 3
a4 = np.array([[10,11,12]]) # 1 3
print(a3 + a4)
```

```python
print(a3 - a4)
```

```python
print(a3 * a4)
```

```python
print(a3 / a4)
```

```python
# 以下为广播各种情况
# (1,2) (1,1)  True
# (2,3) (1,3)  True
# (2,3) (1,4)  False
# (5,2) (5,1)  True
```

```python
a3 = np.array([[1,2,3],[4,5,6]])  # 2 3
a4 = np.array([[10,11]]) # 1 2
print(a3 + a4) # 不满足要求
```

```python
# 对位乘法 表示两个数组的形状是完全一致的
# 矩阵乘法 第一个矩阵的列数 必须 等于 第二个 矩阵的行数
# 最终的结果 为左边的行 以及 右边的列数

a1 = np.array([[1,2,3],[4,5,6]])
a2 = np.array([[6],[4],[2]])


print(a1 @ a2)
```
