## 皮尔逊相关系数

```python
# 1.读取广告数据
# pd.read_csv("../data/advertising.csv")
data = pd.read_csv("/Users/lzc/remote_project/ml_20260331/data/advertising.csv")
print(data.head)
```

```python
# 2.数据清洗
data.drop(data.columns[0], axis=1, inplace=True)  # 删除id列
data.dropna(axis=0, inplace=True)
print(data.head)
```

```python
# 3. 提取特征和标签
x = data.drop("Sales", axis=1, inplace=False)
y = data["Sales"]
print(x.head)
print(y.head)
```

```python
# 4. 计算皮尔逊相关系数
corr = x.corrwith(y, method="pearson")
print(corr)
```

```python
# 5. 相关矩阵
matrix = data.corr(method="pearson")
print(matrix)
```
