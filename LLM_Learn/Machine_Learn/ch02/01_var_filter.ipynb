## 低方差过滤

```python
from sklearn.feature_selection import VarianceThreshold
import numpy as np
```

```python
# 特征1: 标准正态分布  均值0 方差1
x1 = np.random.randn(10000)
x2 = np.random.normal(loc=2, scale=0.1, size=10000)
```

```python
# 构造特征向量
x = np.vstack((x1, x2)).T
print(x)
```

```python
print(f"x1的方差: {np.var(x1)}, x2的方差: {np.var(x2)}")
```

```python
# 低方差过滤
var_thresh = VarianceThreshold(threshold=0.01)
x = var_thresh.fit_transform(x)
print(x)
```
