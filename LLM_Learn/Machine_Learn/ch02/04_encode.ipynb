```python
from uu import encode

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

## 标签编码

```python
data = ["小学", "本科", "中学", "大学", "硕士", "博士"]
```

```python
encoder = LabelEncoder()
data = encoder.fit_transform(data)  # 默认是安装字母表
print(data)
```

## 独热编码

```python
import numpy as np

data = np.array(["red", "yellow", "green", "red", "red", "yellow"]).reshape(-1, 1)
encoder = OneHotEncoder(sparse_output=False)
data = encoder.fit_transform(data)
print(data)
```
