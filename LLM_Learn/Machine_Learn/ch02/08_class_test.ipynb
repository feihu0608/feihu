```python
import numpy as np
import pandas as pd
import sklearn.metrics as metrics
```

## 混淆矩阵

```python
y_true = ["狗", "猫", "猫", "猫", "猫", "猫", "猫", "狗", "狗", "狗"]  # 真实值
y_pred = ["猫", "猫", "猫", "狗", "猫", "猫", "猫", "猫", "狗", "狗"]
labels = ["猫", "狗"]
```

```python
matrix = metrics.confusion_matrix(y_true, y_pred, labels=labels)
print(matrix)
# 方便观察
data = pd.DataFrame(matrix, index=labels, columns=labels)
print(data)
```

## 准确率

```python
print(metrics.accuracy_score(y_true, y_pred))
```

## 精确率

```python
print(metrics.precision_score(y_true, y_pred, pos_label="猫"))
```

## 召回率(查全率)

```python
print(metrics.recall_score(y_true, y_pred, pos_label="猫"))
```

```python
# 平均: 算术平均  几何平均  加权平均  (a1*x1+a2*x2)/(a1+a2)  调和平均  2/f=1/f1+1/f2  => f=2*f1*f2/f1+f2
```

## f1分数 (精确率和召回率的调和平均)

```python
print(metrics.f1_score(y_true, y_pred, pos_label="猫"))
```

## 评估报告

```python
print(metrics.classification_report(y_true, y_pred))
```
