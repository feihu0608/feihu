```python
import numpy as np
```

```python
# 定义一个向量
x = np.array([10, 20, 30, 40])
y = np.array([1, 2, 3, 4])
```

```python
# 向量的转置
print(x.shape)
print(x.T.shape)
```

```python
# 向量和表里相乘
print(x * 10)
```

```python
# 向量相乘: 对位相乘
print(x * y)
```

```python
# 向量点积(内积)
print(x.dot(y))
print(x @ y)
```

```python
# 向量的范数
x = np.array([10, 20, 30, 40, 0, -4])
print(np.linalg.norm(x, ord=0))
print(np.linalg.norm(x, ord=1))
print(np.linalg.norm(x, ord=2))
print(np.linalg.norm(x, ord=np.inf))  # 切比雪芙范数
```
